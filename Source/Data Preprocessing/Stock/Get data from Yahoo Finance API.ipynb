```python
import yfinance as yf
import pandas as pd

interval = "5m"
```

```python
import datetime

today = datetime.date.today()

date_before = today - datetime.timedelta(days=60)

print(date_before)
```

```python
symbol = "AAPL"
data = yf.download(symbol, start="2023-04-03", end="2023-05-31", interval=interval)
data.index = data.index.tz_convert(None)

# Save the data to an Excel sheet
df = pd.DataFrame(data)
df.to_csv("AAPL_data_5min.csv")
```
